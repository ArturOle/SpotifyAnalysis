```python
merged = data_merged.loc[:,["title", "Track Name"]].isnull().values.all(axis=1)
null_indexes = [i for i in range(len(merged)) if merged[i]]
data_merged.loc[null_indexes,:].head(5) # our data is properly merged
```

```python
merged = data_merged.loc[:,["Track Name"]].isnull().values.all(axis=1)
null_indexes = [i for i in range(len(merged)) if merged[i]]
not_null_indexes = [i for i in range(len(merged)) if not merged[i]]
data_merged.loc[null_indexes,:].head(5)
```

```python
data_merged["Track Name"].fillna(data_merged.title, inplace=True)  
data_merged["Artist"].fillna(data_merged.artist, inplace=True)
data_merged["Date"].fillna(data_merged.added, inplace=True)
data_merged["Position"].fillna(data_merged["rank"], inplace=True)
```

#### Setting dates

```python
year = []
month = []
day = []
for date in data_merged["Date"]:
    if pd.isna(date):
        year.append(np.nan)
        month.append(np.nan)
        day.append(np.nan)
    else:
        year.append(date[:4])
        month.append(date[5:7])
        day.append(date[8:10])

date = pd.DataFrame(
    {'year': year,
     'month': month,
     'day': day
    })

data_merged["Date"] = pd.to_datetime(date)
```

```python
data_merged["top year"].fillna(data_merged["Date"].dt.year, inplace=True)
data_merged.drop(columns=['title', 'artist', 'added', "rank"], inplace=True)
data_merged.loc[null_indexes,:].head(5)
```

#### Setting genres

```python
genres = [[], [], []]

for genre_list, genre in zip(data_merged["Genre"], data_merged["top genre"]):
    if pd.isna(genre_list) or "Sorry" in genre_list or "Internal Server Error" in genre_list:
        genres[0].append(genre)
        genres[1].append(np.nan)
        genres[2].append(np.nan)
    else:
        genre_list = ast.literal_eval(genre_list)
        for i in range(min(len(genre_list), 3)):
            genres[i].append(genre_list[i])
        for i in range(len(genre_list), 3):
            genres[i].append(np.nan)
```

```python
data_merged["genre_1"] = genres[0]
data_merged["genre_2"] = genres[1]
data_merged["genre_3"] = genres[2]
```

```python
data_merged.drop(columns=["Genre", "top genre"], inplace=True)
data_merged.loc[null_indexes,:].head(5)
```

```python
artist_genre = pd.concat([
    data_merged.loc[:, ['Artist','genre_1']], 
    data_merged.loc[:, ['Artist','genre_2']].rename(columns={'genre_2':'genre_1'}),
    data_merged.loc[:, ['Artist','genre_3']].rename(columns={'genre_3':'genre_1'})])
artist_genre.dropna(inplace=True)
artist_genre = artist_genre.groupby(['Artist', 'genre_1']).size().reset_index(name="occurences")
artist_genre.head(6)
```

```python
best_genre = artist_genre.loc[artist_genre.groupby('Artist')["occurences"].idxmax(), ["Artist", "genre_1"]]
best_genre.rename(columns={'genre_1':'main_genre'}, inplace=True)
best_genre.head(5)
```

#### SongProps pd df creation

```python
SongProps = data_merged.loc[:, ['Track Name', 'Artist', 
'nrgy', 'dnce', 'dB', 'live', 'bpm', 'val', 'dur', 'acous', 'spch', 'pop', 
'genre_1', 'genre_2', 'genre_3', 'year released', 'artist type', 'top year']]
SongProps.drop_duplicates(ignore_index=True, inplace=True)
SongProps = SongProps.reset_index(level=0)
SongProps.rename(columns = {"index":"song_prop_id", "artist type": "ensemble"}, inplace = True)
SongProps.head(5)
```

```python
SongPropsIdx = SongProps.loc[:, ['song_prop_id', 'Track Name', 'Artist']]
SongProps    = SongProps.loc[:, ['song_prop_id', 'nrgy', 'dnce', 'dB', 'live', 'bpm', 'val', 'dur', 'acous', 'spch', 'pop', 
'genre_1', 'genre_2', 'genre_3', 'year released', 'ensemble', 'top year']]
SongProps.head(5)
```

```python
data_merged = data_merged.merge(SongPropsIdx, how='left', left_on=['Track Name', 'Artist'], right_on=['Track Name', 'Artist'])
data_merged.drop(columns= ['nrgy', 'dnce', 'dB', 'live', 'bpm', 'genre_1', 'genre_2', 'genre_3', 'year released', 
'val', 'dur', 'acous', 'spch', 'pop', 'artist type', 'top year'], inplace=True)
data_merged.head(5)
```

```python
SongProps.to_csv("SongProps.csv", index=False)
```

#### Artist pd df creation

```python
Artists.to_csv("Artists.csv", index=False)
```

```python
data_merged = data_merged.merge(Artists.loc[:, ['artist_name', 'artist_id']], how='left', left_on='Artist', right_on='artist_name')
data_merged.drop(columns= ['artist_name', 'Artist', 'main_genre'], inplace=True)
data_merged.head(5) # potential revenue df
```

#### Price dataframe

```python
Pricing = pd.DataFrame(
    {'year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
     'price_per_stream': [0.005, 0.005, 0.006, 0.007, 0.0045, 0.004, 0.0038, 0.0036, 0.0034, 0.0033, 0.0032, 0.0031, 0.0037],
    })
Pricing.to_csv("Pricing.csv", index=False)
Pricing
```

```python
data_merged["year"] = data_merged["Date"].dt.year
data_merged = data_merged.merge(Pricing, how='left', left_on=['year'], right_on=['year'])
```

```python
data_merged.head()
```

```python
for i in range(data_merged["song_prop_id"].max()):
    streams = data_merged[data_merged["song_prop_id"]==i]["Streams"].array
    print(streams)
    y = np.roll(streams, 1)
    y[0] = y[1]
    streams = streams-y
    print(streams)
    break
```

As we can see, the difference between some of the consecutive days is negative, therefore on the contrary to our previous assumptions the values of the streams are not cumulative, but from each day.

```python
data_merged["profit"] = data_merged["Streams"]*data_merged["price_per_stream"]
```
